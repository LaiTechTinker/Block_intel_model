```python
#Now let's subtract the last_pool timestamp from first pool timestamp to get our duration
import pandas as pd

# Convert timestamp columns to datetime
df['FIRST_POOL_ACTIVITY_TIMESTAMP'] = pd.to_datetime(df['FIRST_POOL_ACTIVITY_TIMESTAMP'], errors='coerce', utc=True)
df['LAST_POOL_ACTIVITY_TIMESTAMP'] = pd.to_datetime(df['LAST_POOL_ACTIVITY_TIMESTAMP'], errors='coerce', utc=True)

# Compute duration in days (difference between last and first)
df['DURATION'] = (df['LAST_POOL_ACTIVITY_TIMESTAMP'] - df['FIRST_POOL_ACTIVITY_TIMESTAMP']).dt.total_seconds() / (24 * 3600)

# Optional: if you prefer hours instead of days, use this instead
# df['DURATION'] = (df['LAST_POOL_ACTIVITY_TIMESTAMP'] - df['FIRST_POOL_ACTIVITY_TIMESTAMP']).dt.total_seconds() / 3600

# Inspect the result
print(df[['FIRST_POOL_ACTIVITY_TIMESTAMP', 'LAST_POOL_ACTIVITY_TIMESTAMP', 'DURATION']].head())


```

### Check Multicollinearity for Categorical features

* **A chi-squared test (also chi-square or 2 test) is a statistical hypothesis test that is valid to perform when the test statistic is chi-squared distributed under the null hypothesis, specifically Pearson's chi-squared test**

* **A chi-square statistic is one way to show a relationship between two categorical variables.**

* **Here we test correlation of Categorical columns with Target column i.e case_status**

Null Hypothesis ($ H_0 $): The Feature is independent of target column (No-Correlation)

Alternative Hypothesis ($ H_1 $): The Feature and Target column are not independent (Correalted)

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in categorical_features:
    if chi2_contingency(pd.crosstab(df['INACTIVITY_STATUS'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[categorical_features, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```
